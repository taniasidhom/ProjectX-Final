```python
import pandas as pd
import os
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

#load in data
tweets_by_hashtag = pd.read_csv('tweets_by_hashtag.csv')
tweets_by_keyword = pd.read_csv('tweets_by_keyword.csv')
```

```python
# only keep hashtag and tweet
hashtags = tweets_by_hashtag.iloc[:, 0:2]
# only keep keyword and tweet
keywords = tweets_by_keyword.iloc[:, 0:2]
```

```python
def sentiment_scores(data):
    tweet_list = data.tweet.tolist()
    analyzer = SentimentIntensityAnalyzer()
    outcome = [analyzer.polarity_scores(tweet) for tweet in tweet_list]
    scored_df = pd.concat([data, pd.DataFrame(outcome)], axis=1)
    return scored_df
```

```python
# sentiment for hashtags
hashtags_sentiment = sentiment_scores(hashtags)
hashtags_sentiment = hashtags_sentiment.rename(columns={'hashtag':'query'})
```

```python
# sentiment for keywords
keywords_sentiment = sentiment_scores(keywords)
keywords_sentiment = keywords_sentiment.rename(columns={'keyword':'query'})
```

```python
#combine both in df 
overall_sentiment = pd.concat([hashtags_sentiment, keywords_sentiment])
overall_sentiment
```

```python
#save all to csvs
keywords_sentiment.to_csv('keywords_sentiment.csv', index=False)
hashtags_sentiment.to_csv('hashtags_sentiment.csv', index=False)
overall_sentiment.to_csv('overall_sentiment.csv', index=False)
```
